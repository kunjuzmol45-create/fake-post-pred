```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('fake_job_postings.csv',encoding='latin-1')
```

```python
df[df['description'].isnull()].head()
```

```python
df.drop(df[df['description'].isnull()].index, inplace = True)
```

```python
df[df['description'].isnull()].head()
```

```python
example="New competition launched :https://www.kaggle.com/c/nlp-getting-started"
```

```python
import re
```

```python
def remove_URL(tweet):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'',tweet)

remove_URL(example)
```

```python
import string
```

```python
def remove_punct(text):
    table=str.maketrans('','',string.punctuation)
    return text.translate(table)

example="I am a #king"
print(remove_punct(example))
```

```python

x_train=df['description']
y_train=df['fraudulent']
```

```python
from sklearn.model_selection import train_test_split
print(x_train.shape)
print(y_train.shape)
```

```python
pd.set_option('display.max_rows', 500)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv= TfidfVectorizer(min_df=1,stop_words='english')
```

```python
X_train_counts=cv.fit_transform(x_train)
```

```python
X_train_counts.toarray()
```

```python
cv.vocabulary_.get(u'algorithm')
```

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier= KNeighborsClassifier(n_neighbors=20, metric='minkowski', p=2 )  
 
classifier.fit(X_train_counts,y_train)
classifier.score(X_train_counts,y_train)
```
